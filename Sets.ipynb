```python
#sets
# fruit={}#empty dictionary
fruit=set()#empty set
print(fruit)
```

```python
fruits={"apple","banana","kiwi","orange"}
print(fruits)
```

```python
fruits.remove("apple")
print(fruits)
```

```python
fruits.add("mango")#adding a single item in your set
print(fruits)
```

```python
fruits.update(["mango","apple"])#adding multiple items
print(fruits)
```

```python
number=set([1,2,3,2,5,1])
print(number)
```

```python
print("cherry" in fruits)#checking membership
```

```python
print(len(fruits))#length of set
```

```python
letter=set("nato")
print(letter)
```
